```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Droped.csv')
data.sort_values(by='Coralation', inplace=True)
data
```

```python
data['Time'] = data['Time'] * 0.0001
```

```python
##clearing some feature for coralation
data_copy_coral = data.copy()
data_copy_coral = data_copy_coral.drop('Droped', axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = data_copy_coral.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
correlation_matrix
```

```python
import seaborn as sns
data_copy_visual = data.copy()
data_copy_visual = data_copy_visual.drop('Droped', axis=1)
data_copy_visual = data_copy_visual.drop('Coralation', axis=1)
data_copy_visual = data_copy_visual.drop('Time', axis=1)

plt.figure(figsize=(10, 8))
sns.heatmap(data_copy_visual, annot=True, cmap='coolwarm', yticklabels=data['Droped'])
plt.title('Heatmap')
plt.show()
```
